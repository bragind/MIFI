```python
#Каков средний балл по чтению у студентов, которые посещали курсы подготовки к экзаменам?
# Фильтрация строк, где test preparation course == 'completed'
filtered_df = student_data.query("`test preparation course` == 'completed'")

# Вычисление среднего балла по чтению
average_reading_score = filtered_df['reading score'].mean()
average_reading_score
```

```python
#Сколько студентов получили 0 баллов по математике?
# Фильтруем строки, где math score равен 0
zero_math_scores = student_data.query("`math score` == 0")

# Количество таких строк
number_of_zero_math_scores = len(zero_math_scores)

print(number_of_zero_math_scores)
```

```python
'''Проверьте гипотезу: у студентов с оплачиваемым питанием средний балл по математике выше, чем у студентов с льготным питанием.
В качестве ответа напишите наибольший средний балл по математике среди этих групп студентов.
Округлите ответ до целого числа.'''
# Группы студентов по типу питания
paid_lunch_students = student_data.query("`lunch` == 'standard'")
reduced_lunch_students = student_data.query("`lunch` != 'standard'")

# Средние баллы по математике для каждой группы
avg_paid_math_score = paid_lunch_students['math score'].mean()
avg_reduced_math_score = reduced_lunch_students['math score'].mean()

# Наибольший средний балл по математике среди двух групп
max_avg_math_score = max(avg_paid_math_score, avg_reduced_math_score)

# Округляем до целого числа
rounded_max_avg_math_score = round(max_avg_math_score)

print(rounded_max_avg_math_score)
```

```python
'''Каков процент студентов, родители которых имеют высшее образование уровня бакалавриата (bachelor's degree)?
Округлите ответ до целого числа.'''
student_data["parental level of education"].value_counts(normalize=True)
```

```python
#Насколько медианный балл по письму у студентов в расовой группе А отличается от среднего балла по письму у студентов в расовой группе C?
# Группа А
group_A = student_data.query("`race/ethnicity` == 'group A'")
median_writing_group_A = group_A['writing score'].median()

# Группа C
group_C = student_data.query("`race/ethnicity` == 'group C'")
mean_writing_group_C = group_C['writing score'].mean()

# Абсолютная разница между медианой группы А и средним группы C
difference = abs(median_writing_group_A - mean_writing_group_C)

# Округление разницы до целого числа
rounded_difference = round(difference)

print(rounded_difference)
```
